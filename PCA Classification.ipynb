```python
def apply_pca(X_train, X_test, n_components):
    pcalist = []
    from sklearn.decomposition import PCA
    pca = PCA(n_components=n_components)
    pcamodellist = [pca]
    for i in pcamodellist:
        print(i)
        X_train_pca = i.fit_transform(X_train)
        X_test_pca = i.transform(X_test)
        explained_variance = i.explained_variance_ratio_
        
        pcalist.append([X_train_pca, X_test_pca, explained_variance])
        
    return pcalist
```
